### Re-classify

```python
def reshape_lime_explanation(X_test, y_test, LIME_explanation):
    explanation = np.repeat(LIME_explanation, X_test.shape[-1]//10).reshape(X_test.shape[0],-1)
    if explanation.shape[-1] != X_test.shape[-1]: #recalibrate LIME explanation
        last_step_explanation = np.transpose(LIME_explanation)[-1].reshape(-1,1)
        n_pad = X_test.shape[-1] - explanation.shape[-1]
        padded_array = np.repeat(last_step_explanation, n_pad, axis=-1)
        explanation = np.append(explanation, padded_array, axis=-1)
    return explanation
```

```python
# Load weights
xais = ['LIME', 'MrSEQL', 'Resnet','GradientShap','IG'] 
names =[ 'lime_mrseql', 'mrseql', 'cam','GS','IG']
weights = []
for xai in xais:
    test_weight_file = 'exp_weights/weights_%s_%s.txt' % (xai,ds)
    weight = np.genfromtxt(test_weight_file, delimiter = ',')
    if xai == 'LIME':
        weight = reshape_lime_explanation(X_test, y_test, weight)
    print(weight.shape)
    weights.append(weight)

for seed in [2020]:
# for seed in [1000,2020,9090,3030,5050]:
    np.random.seed(seed)
    names.append('random'+str(seed))
    random_weights = np.squeeze(np.random.uniform(0,100,size=X_test.shape))
    weights.append(random_weights)
    
print(len(weights))
print(len(weights[0][0]))
print(names)
```
